```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display
import os
```

```python
# Lista de CEDEARs de petroleras y sus nombres completos
cedears_argentinos = {
    "YPFD.BA": "YPF",
    "BMA.BA": "Banco Macro",
    "GGAL.BA": "Grupo Financiero Galicia",
    "EDN.BA": "Edenor",
    "MELI.BA": "MercadoLibre"
}
```

```python
# Obtener datos de cotización diarios
datos_acciones = yf.download(list(cedears_argentinos.keys()), period="5y")

# Precios de cierre
precios_cierre = datos_acciones["Close"]

# Mostrar los datos
print(precios_cierre)

# Guardar los datos en un archivo CSV
# precios_cierre.to_csv("precios_cedears_petroleras.csv")
```

```python
# Función para actualizar el gráfico
def actualizar_grafico(empresa):
    fig.data = []  # Limpiar el gráfico anterior
    fig.add_trace(go.Scatter(x=precios_cierre.index, y=precios_cierre[empresa], name=empresa))
    fig.update_layout(title=f"Tendencia de Precio de {empresa}",
                      xaxis_title="Fecha",
                      yaxis_title="Precio de Cierre (USD)",
                      hovermode="x")
```

```python
# Función para actualizar el gráfico
def actualizar_grafico(empresa_ticker):
    empresa_nombre = cedears_argentinos[empresa_ticker]
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=precios_cierre.index, y=precios_cierre[empresa_ticker], name=empresa_nombre))
    fig.update_layout(title=f"Tendencia de Precio de {empresa_nombre}",
                      xaxis_title="Fecha",
                      yaxis_title="Precio de Cierre ($)",
                      hovermode="x")
    display(fig)

# Crear el widget desplegable
widget_empresas = widgets.Dropdown(options=list(cedears_argentinos.keys()), description="Empresa:")

# Conectar el widget con la función de actualización y llamar a la función inicialmente
widgets.interact(actualizar_grafico, empresa_ticker=widget_empresas)

# Mostrar el gráfico inicial (ya no es necesario aquí)
# display(fig)
```
